```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['feature1', 'feature2']].values  
y = df['price'].values
```

pandas DataFrame iindeki 'feature1' ve 'feature2' stunlarn seer ve bu stunlar bir NumPy dizisine dntrr.

values zellii, DataFrame'in sadece deerlerini (yani stun adlarn ve satr endekslerini darda brakarak) ieren bir NumPy dizisi dndrr. Bu ekilde, seilen stunlar NumPy dizisine dntrlebilir ve daha sonra bir makine renimi modeli gibi NumPy dizileri ile alan ilevlere giri olarak verilebilir.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_test.shape
```

Deep Learning modellerinde, verilerin leklendirilmesi ve normalletirilmesi iin eitli scaler yntemleri kullanlabilir. Hangi scaler ynteminin kullanlmas gerektii, veri setinin zelliklerine, modelin yapsna ve kullanlacak optimizer algoritmasna baldr. Aada yaygn olarak kullanlan scaler yntemlerini aklamak istiyorum:

Min-Max Scaler: Bu yntem, verileri belirli bir arala indirger. Verileri minimum ve maksimum deerleri arasnda belirli bir arala leklendirerek yaplr. Bu yntem, genellikle 0 ile 1 arasnda leklendirme yapmak iin kullanlr.

Standard Scaler: Bu yntem, verilerin ortalamasn 0 ve standart sapmasn 1 olarak leklendirir. Verilerin dalmn merkeziletirerek yaplr. Bu yntem, genellikle verilerin normal dald durumlarda kullanlr.

Robust Scaler: Bu yntem, verilerin ortanca deerini 0 ve yzdelik araln belirli bir arala indirger. Verilerin dalmn merkeziletirerek yaplr. Bu yntem, verilerde aykr deerlerin olduu durumlarda kullanlr.

MaxAbs Scaler: Bu yntem, verilerin maksimum mutlak deerini 1 olarak leklendirir. Bu yntem, zellikle sparse matris verileri iin kullanlr.

Hangi scaler ynteminin kullanlaca, veri setinin zelliklerine ve modelin yapsna baldr. Genellikle, Min-Max scaler veya Standard scaler kullanmak iyi sonular verir. Ancak, farkl scaler yntemleri farkl veri setleri ve modeller iin daha iyi sonular verebilir.

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
help(MinMaxScaler)
```

Yapay sinir a (ANN) modellerinde verilerin leklendirilmesi ve normalletirilmesi, eitim performansn artrmak ve daha iyi sonular elde etmek iin nemlidir. Min-Max normalizasyonu, verileri belirli bir aralkta leklendirmek iin kullanlan bir yntemdir. Bu yntem, verileri minimum ve maksimum deerleri arasnda belirli bir arala indirger. rnein, [0,1] aralna leklendirme yaplabilir.

Min-Max normalizasyonu, verilerin her zellii iin minimum ve maksimum deerleri bulup, her zellii bu minimum ve maksimum deer aralnda yeniden leklendirerek yaplr. Bu sayede, verilerin farkl zellikleri arasndaki deer farkllklar azalr ve daha homojen bir veri kmesi elde edilir.

Bu normalizasyon yntemi, ANN'lerde sklkla kullanlr nk modelin verilerin arasndaki farkllklar daha iyi anlamasna ve daha iyi sonular retmesine yardmc olur.

```python
scaler = MinMaxScaler() # Deep Learning de MinMax kullanmak daha iyidir!
```

```python
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

Sadece girdi verileri (X) leklendirilir, kt verileri (y) leklendirilmez. Bu nedenle, scaler.fit() fonksiyonu yalnzca X_train verilerine uygulanr ve X_train verileri leklendirilir. Daha sonra, renilen lekleme parametreleri (minimum ve maksimum deerler) kullanlarak, X_test verileri de leklendirilir. Y_test ve y_train verilerinin leklendirilmemesinin sebebi, bu verilerin kt deerleri olmasdr ve bu deerlerin, girdi verileri gibi saysal deerler olarak ilenmesi gerektii iin leklendirme ilemine ihtiya duyulmaz.

X_train ve X_test olarak ikiye ayryor ve ardndan scaler.fit() yntemiyle eitiyor. Daha sonra X_train ve X_test veri kmelerini scaler.transform() yntemiyle leklendiriyor. Bu, verilerin zelliklerinin ayn lekte olmasn salayarak daha iyi bir performans elde etmek iin yaplan bir n ilemdir.

## Modelling

```python
pip install keras
```

```python
import tensorflow as tf
```

```python
from keras import models
```

```python
from tensorflow.keras.layers import Dense, Activation
```

```python
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(units = 32, activation = "tanh"))     
model.add(Dense(units = 32, activation = "relu"))
model.add(Dense(units = 32, activation = "tanh"))
model.add(Dense(units = 1)) # regression problemi olduu iin 1 output var. 
```

```python
model.compile(optimizer="rmsprop", loss="mse")

#model kuruldu bu model nasl alacak onu burada belirleriz
#optimizer mincost bulur geriye doru optimize eder. rmsprop veya adam kullanlabilir
# loss loss function regression probleminde genelde mse kullanlr 
```

```python
model.fit(x = X_train, y = y_train, batch_size = 128, epochs = 300) #300 defa tm datay modele sokacak her iterasyonda 32 satr kulland (batch_size 32) 32 satrn ortalama hata deerini Gradientdecent ile parametreleri backpropogation ile gncelledi bir somraki epoc bu deerler zerinden train edilecek btn epoc lar bitene  kadar devam edecek 
#loss deeri  o epoc un en sonuncu deerini verir 
# batch_size iin tek tek verilebilir hzl olur ama doruluk oran azalr, yksek verirsek sre uzar doruluk oran artar.datann tamam olursa ok yava olur fakat ok doru sonular alnr.
# datann bykl ve sreye gre ayarlanr.
```

bir epoch tamamlandnda, tm batch'lerin loss deerlerinin ortalamas, epoch loss deerini oluturur. 

```python
model.summary() # her katmanda ka parametre olduunu verir 
```

```python
model.history.history # model eitilirken kan loss deerlerini verir
```

model.history.history zellii, modelin eitim ilemi srasnda kaydedilen tm deerleri ieren bir dictionary'dir. Bu deerler, eitim srasnda kaydedilen loss ve accuracy (veya baka bir metrik) deerlerini ierir. Bu dictionary, eitim srasnda kaydedilen her epoch iin bir girdiye sahiptir.

pd.DataFrame(model.history.history) kodu, bu dictionary'i Pandas DataFrame formatnda gsterir. Bu kodu altrarak, eitim srasnda kaydedilen her epoch iin loss ve accuracy deerlerini daha dzenli ve okunakl bir ekilde grebilirsiniz. Her bir epoch'un deerleri, DataFrame'deki her bir satrda grntlenir.

```python
pd.DataFrame(model.history.history)
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
```

```python
fig, ax= plt.subplots()
ax.plot(loss_df)
ax.set_xlabel("epoch")
ax.set_ylabel("loss")
ax.set_title("loss - epoch graph")
ax.set_xlim([0, 250]) # focusing on the given x values
ax.set_ylim([0, 3000]) # focusing on the given y values
```

## Model Performance

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score
```

```python
model.evaluate(X_train, y_train, verbose=0) # model olutururken loss'a hangi skoru verdiysem onun sonucunu verir. "mse"
# arka planda x ve y testi predict alp sonra evaluate sonucunu veriyor
```

model.evaluate(X_test, y_test, verbose=0) kodu, eitilen modelin test verileri zerinde performansn deerlendirmek iin kullanlr. Bu yntem, test verileri zerinde modelin loss deerini hesaplar.

```python
model.evaluate(X_test, y_test, verbose=0)
```

```python
model.evaluate(X_test, y_test, verbose=2)
```

```python
y_pred = model.predict(X_test)
```

```python
explained_variance_score(y_test, y_pred) # R2 skorunu verir 1 e yaknsa iyidir
```

Bu kod blou, bir makine renmesi modelinin test verileri zerindeki tahminlerinin doruluunu lmek iin kullanlan bir metrik olan "explained variance score" (aklanan varyans skoru) hesaplamak iin kullanlr.

lk satr, "model.predict(X_test)" kullanarak, X_test veri kmesindeki zelliklere dayanarak modelin y_test verileri iin tahminlerini oluturur. Tahminler, y_pred deikeninde saklanr.

kinci satr, "explained_variance_score" fonksiyonunu kullanarak gerek y_test verileri ve tahmin edilen y_pred verileri arasndaki fark hesaplar. Bu skor, gerek verilerin deikenliini ne kadar iyi akladn gsterir. Skor 0 ile 1 arasnda bir deer alabilir ve ne kadar yksek olursa, modelin tahminleri gerek verilere o kadar yakndr.

```python
my_dict = {"Actual" : y_test, "Pred" : y_pred.reshape(-1,)}
compare = pd.DataFrame(my_dict)
compare.sample(10)

# gerek ve tahmin deerlerini kyaslarz
#byk datalarda  milyonluk datalarda DL modeller tercih edilir 
```

Bu kod blou, test setindeki gerek ve tahmin edilen deerleri karlatrmak iin kullanlr.

lk satr, "my_dict" adl bir szlk oluturur. Bu szlk, iki anahtar deerine sahiptir: "Actual" ve "Pred". "Actual" anahtarnn deeri, y_test adl gerek hedef deikeninin tamamn ierirken, "Pred" anahtarnn deeri, modelin tahminleri olan y_pred deikenini bir boyutlu dizi olarak ierir. "reshape(-1,)" kullanarak y_pred dizisini bir boyut kltr ve bir dizi haline getirir.

kinci satr, "compare" adl bir DataFrame oluturur. Bu DataFrame, "Actual" ve "Pred" stunlarn ierir ve bunlarn deerleri, szlkten alnr. "sample(10)" metodu, DataFrame'den rastgele 10 rnek alarak ve bu rnekleri ekranda gstererek karlatrma ilemini grselletirir.

Bu kod blou, gerek ve tahmin edilen deerler arasndaki farklar gsteren bir tablo oluturur ve bu tablo, modelin performansn deerlendirmeye yardmc olur.

```python
sns.scatterplot(x = 'Actual', y = 'Pred', data = compare)
```

Bu kod blou, "compare" DataFrame'inin "Actual" ve "Pred" stunlarndaki verileri kullanarak, gerek deerler ile tahmin edilen deerler arasndaki ilikiyi grselletirmek iin seaborn ktphanesindeki "scatterplot" fonksiyonunu kullanr.

"scatterplot" fonksiyonu, x ve y eksenlerindeki deikenleri belirterek bir salm grafii izer. "x" parametresi, "Actual" stunundan alnan gerek deerleri, "y" parametresi ise "Pred" stunundan alnan tahmin edilen deerleri temsil eder. "data" parametresi, verilerin bulunduu DataFrame'i belirtir.

Bu kod blou, gerek ve tahmin edilen deerler arasndaki ilikiyi gsteren bir salm grafii oluturur. Eer model doru tahminler yapmsa, salm grafiindeki noktalarn bir izgiye yakn bir ekilde yerleecei beklenir.

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    score = r2_score(actual, pred)
    return print("r2_score:", score, "\nmae:", mae, "\nmse:", mse, "\nrmse:", rmse)
```

```python
eval_metric(y_test, y_pred)
```
