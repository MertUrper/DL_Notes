CNN (Convolutional Neural Network) modelleri olutururken Image Data Generator kullanmak genellikle iyi bir uygulamadr. Image Data Generator, zellikle byk veri kmeleri zerinde alrken ve veri artrma (data augmentation) ilemleri yapmak istediinizde ok faydaldr. te Image Data Generator'n baz avantajlar:

Veri Artrma: Veri Artrma, mevcut eitim veri kmenizi eitlendirmek ve modelinizi daha genelletirilebilir hale getirmek iin kullanlan bir tekniktir. Image Data Generator, resimleri rastgele dndrme, yeniden boyutlandrma, yatay veya dikey olarak aynalama, parlaklk deiimi gibi ilemlerle artrabilir. Bu, modelinizin daha fazla eitlilikle eitilmesini salar.

Veri Ykn Azaltma: Image Data Generator, bellek verimliliini artrmak iin resimleri ykleme ilemi srasnda gerek zamanl olarak dntrebilir. Bu, byk veri kmeleriyle alrken ilemci ve bellek kaynaklarn daha etkili bir ekilde kullanmanza yardmc olur.

Mini-Ynlar (Mini-Batches) Oluturma: Image Data Generator, eitim srasnda veriyi mini-ynlar halinde (mini-batches) otomatik olarak oluturabilir. Bu, byk veri kmelerinin daha iyi ilenmesini salar ve ayn zamanda eitim srecini hzlandrabilir.

Kolay Kullanm: Image Data Generator, Keras ve TensorFlow gibi yaygn kullanlan derin renme ereveleriyle uyumlu bir ekilde kullanlabilir. Bu, modelinizi olutururken ve eitirken kolayca entegre edilebilir.

Ancak Image Data Generator' kullanrken dikkate almanz gereken baz nemli hususlar vardr. rnein, veri artrma ilemleri, veriye zg olarak yaplmaldr ve baz durumlarda fazla veri artrma modelinizi ktletirebilir. Ayrca, baz uygulamalarda Image Data Generator' kullanmadan nce resimleri elle n ilemeniz gerekebilir.

Sonu olarak, Image Data Generator, CNN modelleri olutururken veri artrma ve veri ykn azaltma ilemleri iin gl bir aratr ve genellikle iyi bir uygulamadr. Ancak veri kmenizin zelliklerine ve projenizin gereksinimlerine bal olarak kullanmn dikkatlice planlamalsnz.

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#help(ImageDataGenerator)
```

ImageDataGenerator snf, Keras ktphanesinde yer alan ve grnt verilerini artrarak daha fazla veri retmek iin kullanlan bir veri artrma aracdr.

Bu kod blou, ImageDataGenerator snfndan bir rnek oluturur ve bu rnee veri artrma yntemlerinin parametrelerini belirtir. Aadaki parametreler kullanlr:

rotation_range: Rastgele dndrme as aral (0-180 derece arasnda).

width_shift_range: Rastgele yatay kaydrma aral, grnt geniliinin yzdesi olarak belirtilir.

height_shift_range: Rastgele dikey kaydrma aral, grnt yksekliinin yzdesi olarak belirtilir.

shear_range: Rastgele kayma aral, kayma as derecesi olarak belirtilir.

zoom_range: Rastgele yaknlatrma aral.

horizontal_flip: Rastgele yatay evirme yapar.

fill_mode: Grnt geniletilirken veya kltlrken oluan bo piksellerin doldurulma yntemi. 'nearest' yntemi, piksellerin en yakn komusunun rengi ile doldurulmasn salar.

Bu parametreler, image_gen adl bir veri artrma nesnesi oluturur ve daha sonra modelin eitimi srasnda bu nesne kullanlarak veri artrma yaplabilir. Bu, modelin daha fazla veriye erimesine ve genelletirme yapmasna yardmc olur.

```python
image_gen = ImageDataGenerator(rotation_range=15, 
                               width_shift_range=0.10, 
                               height_shift_range=0.10, 
                               #rescale=1/255,   # resimler ncesinden rescale edilmi olduundan tekrar rescale yapmadk.
                               shear_range=0.1, 
                               zoom_range=0.1, 
                               horizontal_flip=True, 
                               fill_mode='nearest')
```


ImageDataGenerator ile veri artrma ilemi, var olan veri kmesindeki grntleri maniple ederek yeni grntler oluturur. Bu nedenle, eitim veri setindeki veri says artmaz, ancak modelin eitim srasnda kullanabilecei daha fazla rnek varm gibi davranr.

rnein, eer eitim veri setinde 1000 adet grnt varsa ve ImageDataGenerator' belirtilen parametrelerle kullanarak her grntden 10 yeni grnt oluturulursa, toplam eitim verisi says 10000 olur. Ancak bu yeni verileri diskte depolamak yerine, modelin eitimi srasnda dinamik olarak retilir ve bellekte tutulur. Bu nedenle, veri artrma ilemi eitim verisi saysn artrmaz, ancak modelin eitim srasnda daha fazla veriye erimesini salar.

# Creating the Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense, Conv2D, MaxPooling2D
```

```python
model = Sequential()

model.add(Conv2D(filters=64, kernel_size=(3,3),input_shape=image_shape, padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu',))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu',))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=64, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))


model.add(Flatten())


model.add(Dense(128))
model.add(Activation('relu'))

model.add(Dropout(0.5))

model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.summary()
```

## Early Stopping

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience=10)
```

## Training the Model

```python
#help(image_gen.flow_from_directory)
```

```python
batch_size = 32
train_image_gen = image_gen.flow_from_directory(train_path,
                                               target_size=(130,130),
                                               color_mode='rgb',
                                               batch_size=batch_size,
                                               class_mode='binary', shuffle=True, seed=42)
```

image_gen.flow_from_directory() yntemi, verilerin diskteki bir dizindeki alt dizinlerde snflandrlm olduu bir klasr yapsn kullanarak, eitim verileri iin bir veri ak oluturur.

Bu yntemin parametreleri yledir:

directory: Eitim verilerinin bulunduu dizin yolu.
target_size: Grntlerin yeniden boyutlandrlaca hedef boyut.
color_mode: Grnt renk kanallarnn modu, rn. "rgb" veya "grayscale".
batch_size: Tek seferde oluturulacak veri ynnn boyutu.
class_mode: Snflandrma yntemi, rn. "binary" veya "categorical".
shuffle: Verilerin kartrlp kartrlmayaca.
seed: Verilerin kartrlmasnda kullanlacak rastgele say retecinin nceden belirlenmi bir balang deeri.
Bu yntem, her arldnda bir "batch" veri retir, bylece eitim srasnda modelin her admda yeni verilerle almas salanr. Ayrca, ImageDataGenerator ile belirlenen veri artrma ilemleri, her arda dinamik olarak uygulanr, bylece eitim srasnda model daha fazla veriye eriir ve daha iyi genelletirme yapabilir.

```python
test_image_gen = image_gen.flow_from_directory(test_path,
                                               target_size=(130,130),
                                               color_mode='rgb',
                                               batch_size=batch_size,
                                               class_mode='binary', shuffle=True, seed=42)
```

image_gen.flow_from_directory() yntemi, verilerin diskteki bir dizindeki alt dizinlerde snflandrlm olduu bir klasr yapsn kullanarak, test verileri iin bir veri ak oluturur.

Bu yntemin parametreleri yledir:

directory: Test verilerinin bulunduu dizin yolu.
target_size: Grntlerin yeniden boyutlandrlaca hedef boyut.
color_mode: Grnt renk kanallarnn modu, rn. "rgb" veya "grayscale".
batch_size: Tek seferde oluturulacak veri ynnn boyutu.
class_mode: Snflandrma yntemi, rn. "binary" veya "categorical".
shuffle: Verilerin kartrlp kartrlmayaca.
seed: Verilerin kartrlmasnda kullanlacak rastgele say retecinin nceden belirlenmi bir balang deeri.
Bu yntem, her arldnda bir "batch" veri retir, bylece test srasnda modelin her admda yeni verilerle almas salanr. Ayrca, ImageDataGenerator ile belirlenen veri artrma ilemleri, test verilerinde uygulanmaz, nk test verileri doru sonularla karlatrmak iin orijinal hallerinde kalmas gereklidir.
